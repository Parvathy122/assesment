```python
#a.
df=pd.read_csv('/diabetes_model.csv')
subset=df.loc[45:60,['Glucose','Age']]
print("\nGlucose level and age for samples 45 to 60:")
subset
```

```python
#b.
bp=df[df['BloodPressure']>120].shape[0]
print("\nNo of patients with blood pressure above 120:",bp)
```

```python
import matplotlib.pyplot as plt
```

```python
#c.
min=df['Insulin'].min()
max=df['Insulin'].max()
print("\nMaximum insulin level:",max)
print("\nMinimum insulin level:",min)
```

```python
#d.
plt.figure(figsize=(10,6))
plt.hist(df['Age'],bins=20,color='blue',edgecolor='black')
plt.title("Histogram of Age ")
plt.xlabel("'Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#e.
plt.figure(figsize=(10,6))
plt.scatter(df['BMI'],df['Glucose'],alpha=0.5,color='green')
plt.title("Scatter plot of BMI vs Glucose")
plt.xlabel("BMI")
plt.ylabel("Glucose")
plt.show()
```
